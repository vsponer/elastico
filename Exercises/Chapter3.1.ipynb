# Euler Method

```python
import numpy as np
import math
import matplotlib.pyplot as plt

#r0 = input("Enter initial radial distance = ")      #in (AU) -> NEED TO CONVERT ALL INPUTS IN INTEGERS OR FLOAT
r0 = 1
r = [r0, 0]                                         #initial position is on x-axis
#v0 = input("Enter initial tangential velocity = ")  #in (AU/yr)
v0 = 2*math.pi
v = [0, v0]                                         #initial velocity is in the y-direction
#tau = input("Enter time step, tau = ")              #in (yr)
tau = 0.02

GM = 4*math.pi**2                                   #Gravitational constant * Mass of sun (AU^3/yr^2)
mass = 1                                           #mass of projectile

#MAIN LOOP
time = 0
nstep = 200                                         #Number of time steps
rplot = np.zeros(nstep)
thetaplot = np.zeros(nstep)
timeplot = np.zeros(nstep)
kinetic = np.zeros(nstep)
potential = np.zeros(nstep)
for n in np.array(range(nstep)):
    normr = math.sqrt(r[0]**2+r[1]**2)              #calculate the to norm -> trying to find norm(x) function
    normv = math.sqrt(v[0]**2+v[1]**2)
    rplot[n] = normr                                #Record orbit for polar plot 
    thetaplot[n] = math.atan2(r[1],r[0])            #Record theta for polar plot
    timeplot[n] = time                              #Record time for plot
    kinetic[n] = 0.5*mass*normv**2                  #Record energies      
    potential[n] = -GM*mass/normr  
    #Calculate new position and velocity
    a0 = -GM*r[0]/normr**3
    a1 = -GM*r[1]/normr**3
    r[0] = r[0] + tau*v[0]                          #Euler step
    r[1] = r[1] + tau*v[1]
    v[0] = v[0] + tau*a0
    v[1] = v[1] + tau*a1
    time = time + tau

#Graph the trajectory of the comet
plt.figure(1)
#plt.subplot(121)
plt.polar(thetaplot,rplot, '+')     #creating a polar plot
plt.grid                            #Including grid into the plot
plt.title('Orbital Motion')
plt.ylabel('Distance (AU)')

plt.figure(2)
#plt.subplot(122)
totalE = kinetic + potential        #total energy
plt.plot(thetaplot,kinetic, '-.',thetaplot,potential, '--',thetaplot,totalE, '-')   #plotting all energies LOOK THE PLOTTING SYMBOLS UP
plt.grid                            #Including grid into the plot
plt.title('Kinetic E (dot); Potential E (dash); Total E (solid)')
plt.ylabel('Energy')
plt.xlabel('Time (yr)')
#plt.subplot(111)
```
